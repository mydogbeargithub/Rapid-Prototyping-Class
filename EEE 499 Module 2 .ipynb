## What is an array?
<br> An array is a collection of objects. 
<br> Objects in same aray must be the same type. 
<br> Multidimensional arrays are standard. 
<br> Arrays are denoted by square brackets. 
<br> Array entries are separated by spaces. 
<br> Entries are indexed, starting with 0. 
<br> Python is optimized for arrays. 
<br> Standard operators work entry by entry. 
<br> Functions can work entry by entry. 
<br> Once created, arrays are fixed in size. 
<br> len(an_array) returns  the numbers of rows. 
<br> We will use numpy arrays.


```python
import numpy as np
array_0=np.zeros(4,float)
array_1=np.ones(4,complex)
array_2=np.full([2,3], 23) #specify initial value for all entries
array_3=np.full_like(array_2,2) #zero_like and one_like work the same way, 
#same dimensions as array_2 filled with 2

print(array_0)
print(array_1)
print(array_2)
print(array_3)
```

```python
import numpy as np
array_0 = np.zeros(4,int)
array_1 = np.zeros(4,int)

for index in range(4): 
    array_0[index] = index
    array_1[index] = (index *2) +1
    
print(array_0)
print(array_1)

array_3=array_0*array_1 # element by element multiplication
print(array_3)

array_4=np.zeros([1,4],int)
array_5=np.zeros([4,1],int)
for index in range(4): 
        array_4[0,index] = 1+index
        array_5[index,0] = (index+1) **2

print(array_4)
print(array_5)

my_dot=np.dot(array_4,array_5) #form dot product
print(my_dot)

my_dot=np.dot(array_5,array_4) #very different result when you reverse the dot
#product
print(my_dot)

```

## Exception Catching
<br> Exceptions cause programs to crash
<br> Exceptions can be due to bad code Fix before shipping!
<br> Exceptions can be due to bad data Python can catch these!
<p> Examples: 
<br> ZeroDivision Error, TypeError, RecursionError

```python
try: 
    bad_div = 12/0 #can be a full block of code
except: 
    print("don't do that!")
    
try: 
    bad_div = 12/0 #can be a full block of code
except ZeroDivisionError: 
    print("ZeroDivisionError - still don't do that!")
except: 
    print("some other exception")
    
try: 
    bad_div = 12/0 #can be a full block of code
except TypeError: 
    print("TypeError - still don't do that!")
except: 
    print("some other exception")
```

```python
my_string = "Once upon a time"
print("this is my_string:", my_string)
my_words = my_string.split()
print("my_words is a list:", my_words)
print("my string did not change:", my_string)

for word in my_words: 
    print("iterate on list items:", word)

for index in range(len(my_words)): 
    print("iterate by index:", my_words[index])

str1='what a fine day this is'
print(str1.split())
print(str1.split('f'))
print(str1.split('i'))

delim = '@'                    #use this as the joining character
my_tuple = ('a', 'b', 'c')
my_list = ['d', 'e', 'f']
print(delim.join(my_tuple))
print(delim.join(my_list))

delim=''                       # don't like anything in between
print(delim.join(my_tuple))
print(delim.join(my_list))

delim=' '                       #put a space in between
print(delim.join(my_tuple))
print(delim.join(my_list))
```

```python
my_inp_file = open("m2_str_io.txt", "r") # r is default, add a b for binary
my_out_file = open("m2_str_io.out", "w") # to append, use "a" instead of "w"

for line in my_inp_file:                #neat way to get a line at a time to 
    print(line)                         #to process, note extra carriage return
    my_out_file.write(line)             #print always adds carriage return
    
my_inp_file.close()                      # always close when done
my_out_file.close()                      #forgetting to close is a common error
```
